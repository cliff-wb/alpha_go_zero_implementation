```python
from matplotlib import *
from matplotlib.pyplot import *
import numpy as np
from PIL import Image
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
%matplotlib inline

model_dir = '../models/'

def simpleaxis(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

def train_hist(f, new_fig=True, color=[0,1,0], gnuf=.3, a=.5):
    color = np.asarray(color)
    z = np.load(model_dir + f, allow_pickle=True).item()
    if new_fig:
        sz = 1.5; figure(figsize=(15*sz,4*sz))
    
    n_rows, n_cols = 2, 4
    
    plot_ind = 1
    n_plot = len(z['log']['score_nn'])
    xs = z['log']['opt_batch']
    xse = z['log']['eval_batch']
    
    #### score
    ax = subplot(n_rows,n_cols,plot_ind); plot_ind += 1; simpleaxis(ax)
    plot(xse, z['log']['score_nn'], color=color, alpha=a)
    plot(xse, z['log']['score_nn_gnu'], linewidth=2, color=color*gnuf, alpha=a)
    axhline(0, color='k', alpha=.25)
    title('Score')
    ylabel('Average score')
    legend(['random opponent','GNU Go opponent'])
    ax.set_xticklabels([])
    
    #### captures
    ax = subplot(n_rows,n_cols,plot_ind); plot_ind += 1; simpleaxis(ax)
    plot(xse, z['log']['n_captures_nn'], color=color, alpha=a)
    plot(xse, z['log']['n_captures_opp_nn'], color=color, linestyle='--', alpha=a)
    plot(xse, z['log']['n_captures_nn_gnu'], linewidth=2, color=color*gnuf, alpha=a)
    plot(xse, z['log']['n_captures_opp_nn_gnu'], linewidth=2, color=color*gnuf, linestyle='--', alpha=a)
    axhline(0, color='k', alpha=.25)
    title('Captures')
    ylabel('Average # of captures')
    ax.set_xticklabels([])
    
    #### win
    ax = subplot(n_rows,n_cols,plot_ind); plot_ind += 1; simpleaxis(ax)
    plot(xse, z['log']['win_nn'], color=color, alpha=a)
    plot(xse, z['log']['win_nn_gnu'], linewidth=2, color=color*gnuf, alpha=a)
    axhline(0.5, color='k', alpha=.25)
    title('Win rate')
    ylabel('Average win rate')
    ax.set_xticklabels([])
    
    keys = ['pol_cross_entrop', 'val_mean_sq_err', 'pol_max', 'val_pearsonr']
    titles = ['Policy cross entropy loss', 'Value mean squared error loss', 'Max policy output', 
              'Predicted vs. actual game scores']
    ylabels = ['Cross entropy', 'Mean squared error', 'Policy output', 'Negative Pearson correlation']
    for key, title_txt, ylabel_txt in zip(keys, titles, ylabels):
        ax = subplot(n_rows,n_cols,plot_ind); plot_ind += 1; simpleaxis(ax)
        plot(xs, z['log'][key], color=color, alpha=a)
        title(title_txt)
        ylabel(ylabel_txt)
        xlabel('Training steps')

    suptitle('N_SIM %i   N_TURNS %i' % (z['N_SIM'], z['N_TURNS']))
    return z

psz = 16
pszh = psz / 2
pieces = np.asarray(Image.open('go_pieces.png').resize((psz*2,psz)))[:,:,:3]
piece = [None]*2
piece[1] = pieces[:,:psz] # black
piece[0] = pieces[:,psz:] # white
blank = np.asarray(Image.open('go_blank.png').resize((psz,psz)))[:,:,:3]

# return board as image
def rboard(board):
    n_rows = n_cols = board.shape[0]
    b = np.zeros((psz*n_rows, psz*n_cols, 3), dtype='uint8')
    for i in range(n_rows):
        for j in range(n_cols):
            b[i*psz:(i+1)*psz][:,j*psz:(j+1)*psz] = blank

    for i in range(n_rows):
        b[i*psz + pszh,:] = 1

    for i in range(n_cols):
        b[:, i*psz + pszh] = 1

    for i in range(n_rows):
        for j in range(n_cols):
            if board[i,j] == 0:
                continue
            b[i*psz:(i+1)*psz, j*psz:(j+1)*psz] = piece[1+board[i,j]/2]
    return b

def sboard(board):
    imshow(rboard(board)); axis('off');

def show_game_hist(gm=0, n_cols=10, x_sz=23, y_sz=9):
    keys = ['nn', 'nn_gnu']
    nms = ['Random opponent:', 'GNU Go opponent:']
    n_turns = z['boards']['nn'].shape[0]
    n_rows_per_gm = np.ceil(np.single(n_turns)/n_cols)
    n_rows = len(keys) * n_rows_per_gm
    
    plot_ind = 1
    figure(figsize=(x_sz, y_sz))
    for nm, key in zip(nms, keys):
        if plot_ind != 1:
            plot_ind = n_rows_per_gm*n_cols + 1
        for turn in range(n_turns):
            ax = subplot(n_rows, n_cols, plot_ind); plot_ind+=1
            if turn == 0:
                ax.annotate('%s\n(Score: %i)' % (nm, z['scores'][key][gm]), xy=(0, 0), xytext=(0, 0),
                        xycoords=ax.yaxis.label, textcoords='offset points', size=20, ha='center', va='center')
            sboard(z['boards'][key][turn,gm])
            title('Turn %i' % turn)
    suptitle('Two example game histories playing against a random opponent and GNU go', fontsize=22, fontweight='bold')

def show_end_games(n_turns=20):
    keys = ['nn', 'nn_gnu']
    nms = ['Random opponent:', 'GNU Go opponent:']
    n_rows = len(keys)
    n_cols = n_turns
    plot_ind = 1
    figure(figsize=(23,9))
    for nm, key in zip(nms, keys):
        for c in range(n_cols):
            ax = subplot(n_rows*2,n_cols/2,plot_ind); plot_ind+=1
            if c == 0:
                ax.annotate(nm, xy=(0, 0), xytext=(0, 0),
                        xycoords=ax.yaxis.label, textcoords='offset points', size=20, ha='center', va='center')
            sboard(z['boards'][key][-1,c])
            title('Score: %i' % z['scores'][key][c])
    suptitle('Example terminal game states against a random opponent and GNU go', fontsize=22, fontweight='bold')
    
def show_self_play_parallel_run(batch_set=0, gm=0, n_turns=35):
    plot_ind = 1
    n_row = 3
    n_col = np.ceil(35./n_row)
    a = 3
    figure(figsize=(12*a,2.5*a))
    board = np.array(np.frombuffer(z['board'], dtype='float16'), dtype='int')
    board = board.reshape((np.prod(board.shape)/(7*7*3)), 7,7,3)
    for turn in range(n_turns):
        for player in [0]:
            subplot(n_row, n_col, plot_ind); plot_ind += 1
            title('%i' % turn)
            sboard(board[batch_set*n_turns*128*2 + turn*128*2 + player*128 + gm,:,:,0])
    suptitle('Network self-play game used for training the network', fontsize=22, fontweight='bold')
```
